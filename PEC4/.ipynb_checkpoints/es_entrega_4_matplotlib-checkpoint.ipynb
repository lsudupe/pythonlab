### Ejercicio 1

Representa en un único gráfico las funciones seno y arcoseno en el intervalo [-180.0, 180.0]

```python
%matplotlib inline

# Importamos las librerias correspondientes
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Creamos el array de [-180.0, 180.0] con intervalos de 0.1
x = np.arange(-180.0, 180.0)
a = np.arange(-1, 1, 0.1)   #El arcoseno solo comprende valores entre -1 y 1
# Creamos la representación con ambas funciones
plt.plot(x, np.sin(x), 'r', a, np.arcsin(a), 'g')

# Ponemos los nombres a los ejes
plt.xlabel('x')
plt.ylabel('sen(x) (red) and arcsin(x) (green)')



# Mostramos el gráfico
plt.show()
```
